## Загрузка и подготовка реестра компании

- Считываю данные с помощью pandas
- Вывожу инфорацию по содержимому, чтобы оценить данные

```python
import pandas as pd

df = pd.read_csv('reestr proektov.csv')
df.head()
```

```python
df.info()
```

- Удаляю столбцы, в которых не содержится ключевая информация для анализа - '№', 'Unnamed: 9', 'Тема письма', 'Спорный проект', 'Примечания:', 'Номер задачи в Битриксе'

```python
df = df.drop(columns=['№', 'Unnamed: 9', 'Тема письма', 'Спорный проект', 'Примечания:', 'Номер задачи в Битриксе'])
df.head()
```

```python
df.info()
```

- Переименую столбцы на английские значения для удобства последующей обработки данный

```python
df.rename(columns={'Менеджер': 'Menager'}, inplace=True)
df.rename(columns={'Статья расходов': 'Expense_item'}, inplace=True)
df.rename(columns={'Наименование проекта': 'Project_name'}, inplace=True)
df.rename(columns={'Наименование позиции по прайсу': 'Price_position'}, inplace=True)
df.rename(columns={'Дата постановки задачи': 'Date_start'}, inplace=True)
df.rename(columns={'Цена за ед.(с НДС)': 'Price'}, inplace=True)
df.rename(columns={'Количество (шт.)': 'Quantity'}, inplace=True)
df.rename(columns={'Стоимость': 'Total_cost'}, inplace=True)
df.rename(columns={'Дата сдачи': 'Date_finish'}, inplace=True)
df.rename(columns={'Исполнитель': 'Executor'}, inplace=True)
df.info()
```

- Удаляю строки, если в столбцах 'Total_cost' и 'Date_finish' отсутствуют значение тк вероятно это не завершенные проекты (статистика по которым нам пока не требется) или пустые строки, не нужные для аналитики

```python
df = df.dropna(subset=['Total_cost', 'Date_finish'])
df.head()
```

- Анализирую типы даннных по столбцам и преобразую тип данных столбцов 'Date_start' и 'Date_finish' в дату, а 'Quantity' в число

```python
df['Date_finish'] = pd.to_datetime(df['Date_finish'], format='%d.%m.%Y')
df['Date_start'] = pd.to_datetime(df['Date_start'], format='%d.%m.%Y')
df['Quantity'] = df['Quantity'].astype(int)
df.info()
```

- Добавляю новый столбец с расчетом длительности выполнения проекта в днях. Поскольку день постановки задачи также считается днем выполнения - прибавляем +1 день ко всем расчетам

```python
df['days_count'] = (df['Date_finish'] - df['Date_start']).dt.days + 1
df.head()
```

- Привожу значения в столбце к корректным, заменяем отрицательные значения на 1

```python
df['days_count'] = df['days_count'].apply(lambda x: 1 if x < 0 else x)
df.head()
```

- Столбец "Project_name" имеет слишком длинный текст, это может мешать дальнейшей обработке инфорации. Преобразую тип данных в столбце в строковый и сокращаю количество текста в столбце до 55 символов. 

```python
df['Project_name'] = df['Project_name'].astype(str)
max_length = 55
df['Project_name'] = df['Project_name'].apply(lambda x: x[:max_length])
df.head()
```

## Выгрузка подготовленного документа для дальнейшей работы
